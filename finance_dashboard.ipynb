```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load dataset
df = pd.read_csv("finance_data.csv")
df.head()

# Step 3: Explore data
print(df.info())
print(df.describe())

# Step 4: Category-wise total spending (Bar chart)
category_total = df.groupby("Category")["Amount"].sum()
plt.figure(figsize=(8,5))
category_total.plot(kind="bar", color="skyblue")
plt.title("Total Spending by Category")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.show()

# Step 5: Expense distribution (Pie chart)
plt.figure(figsize=(8,8))
category_total.plot(kind="pie", autopct="%1.1f%%", colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'])
plt.title("Expense Distribution")
plt.ylabel("")
plt.show()

# Step 6: Monthly spending trend (Line chart)
monthly_spending = df.groupby("Month")["Amount"].sum()
plt.figure(figsize=(8,5))
monthly_spending.plot(kind="line", marker="o", color="green")
plt.title("Monthly Spending Trend")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()

# Step 7: Savings vs Shopping comparison (Bar chart)
comparison = df[df["Category"].isin(["Savings","Shopping"])]
pivot_data = comparison.pivot(index="Month", columns="Category", values="Amount")
pivot_data.plot(kind="bar", figsize=(8,5))
plt.title("Savings vs Shopping")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()

# Step 8: Category trends over months
category_month = df.pivot_table(values="Amount", index="Month", columns="Category", aggfunc="sum")
category_month.plot(figsize=(10,6), marker="o")
plt.title("Monthly Spending by Category")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()

# Step 9: Key insights
print("Insights:")
print("- Savings had the highest allocation")
print("- Shopping expenses increased over time")
print("- Entertainment spending was lowest")
plt.savefig("images/expense_distribution.png")

```
